# Importing libraries

```python
import pandas
import csv
```

# Loading data

```python
spotify_user_track = pandas.read_csv(
    "./data/spotify_dataset.csv",
    dtype={"user_id": "string", ' "playlistname"': "category"},
    quoting=csv.QUOTE_ALL,
    on_bad_lines="warn",
    usecols=["user_id", ' "playlistname"'],
)

spotify_user_track
```

```python
spotify_user_track.where(spotify_user_track['playlist_name'] == 'Everything at once').groupby("user_id").count()
```

# Data cleaning

```python
spotify_user_track = spotify_user_track.rename(columns={' "playlistname"': "playlist_name"})

user_playlist_pair = spotify_user_track.drop_duplicates(["user_id", "playlist_name"])
user_playlist_pair
```

```python
user_playlist_user_joined = user_playlist_pair.join(user_playlist_pair, lsuffix="_from", rsuffix="_to")
user_playlist_user_joined.loc[user_playlist_user_joined['user_id_from'] != user_playlist_user_joined['user_id_to']]
```
